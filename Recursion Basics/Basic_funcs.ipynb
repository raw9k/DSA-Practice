# Recursion

```python
# print a name 4 times
# Head Recursion

def name(i):
    if i ==4:
        return
    print("Rounak")
    i +=1
    name(i)
       
name(0)
```

## Printing numbers From 1 to N and vice - Versa

```python
# head Recursion
def printNos(n,i):
    if i ==n:
        return
    i +=1
    print(i ,end = " ")
    printNos(n,i)
printNos(10,0)
```

```python
def printNos(n):
    if n ==0:
        return
    
    print(n ,end = " ")
    printNos(n-1)
printNos(10)
```

## Sun Of N

```python
# head recursion
def sumofn(sum,i,n):
    if i>n:
        print(sum)
        return
    sumofn(sum+i, i+1,n)
    
sumofn(0,1,10)  
```

```python
# Tail Recursion 
def sumofnum(n):
    if n == 0:
        return 0
    return n + sumofnum(n - 1)  

print(sumofnum(10))  
```

## Reversing an array

```python
# reverse an array

array = [1,4,2,5,6]
def reverse(array,l,r):
    if array[l] == array[r]:
        print(array[r], end=" ")
        return
    reverse(array,l+1,r)
    print(array[l], end=" ")
reverse(array,0,-1)

# This approach give output in object type but we need in array
```

```python
# reverse an array
array = [1,4,2,5,6]
def reverse(array,l,r):
    if l==r:
        return array
    array[l],array[r]= array[r],array[l]
    reverse(array,l+1,r-1)
    return array
reverse(array,4,4)
# Now, This gives output in array
```

```python
# Using While Loop
num = [5,7,3,2,6,1,5,9]
l ,r = 0, 7

while l < r:
    num[l] , num[r] = num[r], num[l]
    l , r = l + 1 , r - 1

print(num)
```

## Factorial of N numbers

```python
# LINK: https://www.geeksforgeeks.org/problems/factorial5739/1
class Solution:
    def factorial(self, n: int) -> int:
        if n ==1:
            return 1
        return n*self.factorial(n-1)
    
output = Solution()
output.factorial(5)
```
